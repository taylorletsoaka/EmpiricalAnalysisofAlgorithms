```python
import matplotlib.pyplot as plt
```

```python
worst = "bubbleworstcase.csv"
best = "bubblebestcase.csv"
avg = "bubbleaveragecase.csv"

title = "Bubble Sort Analysis"

plt.plotfile(avg, ('input', 'time'), newfig = False, label = "Average")
plt.plotfile(best, ('input', 'time'), newfig = False, label = "Best")
plt.plotfile(worst, ('input', 'time'), newfig = False, label = "Worst")

plt.xlabel('input size (n)')
plt.ylabel('time (ms)')
plt.title(title)
plt.legend()
plt.savefig(title +'.pdf')



plt.xlim(xmin=0)
plt.savefig(title +'.pdf')
```
