```python
import matplotlib.pyplot as plt
```

```python
best = "bestcase.csv"
avg = "avgcase.csv"
worst = "worstcase.csv"
title = "Linear Search Analysis"

plt.plotfile(best, ('input', 'time'), newfig = False, label = "Best")
plt.plotfile(avg, ('input', 'time'), newfig = False, label = "Average")
plt.plotfile(worst, ('input', 'time'), newfig = False, label = "Worst")

plt.xlabel('input size (n)')
plt.ylabel('time (ms)')
plt.title(title)
plt.legend()
plt.xlim(xmin=0)
plt.savefig(title +'.pdf')
```
